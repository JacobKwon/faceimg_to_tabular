```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

#global
TotalAngleDataLists = np.empty((0,2), float)

FACIAL_LANDMARKS_TOTAL_IDXS = OrderedDict([
	("right_eyebrow", (17, 22)),
	("left_eyebrow", (22, 27)),
	("right_eye", (36, 42)),
	("left_eye", (42, 48)),
	("nose", (27, 36)),
	("mouth", (48, 68)),
	#("inner_mouth", (60, 68)),
	("jaw", (0, 17))
])
```

```python

def folder_read(folderName):
    
    if( os.path.isdir(folderName) == True):
        f_list = os.listdir(folderName)
        
        for idx, img_name in enumerate(f_list):
            print(img_name)
            face_data_with_angle(folderName+'/'+img_name, detector, predictor, idx)
    
    else:
        print('Not Search '+folderName+' folder.')
    
```
